# Feature selection

## Encoding Target

```python
le = LabelEncoder()
big_df['Course_Encoded'] = le.fit_transform(big_df['Course'])
```

```python
big_df.head(3)
```

## Target Fetch

```python
y=big_df['Course_Encoded']
```

```python
y.head(2)
```

## Features fetch

```python
big_df['combined_text'] = big_df['Interest'] + " " +big_df['Career Goal'] + " " + big_df['Skills']
```

```python
x=big_df[['combined_text', 'Budget']]
```

```python
x.shape
```

```python
y.shape
```

## Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)
```

```python
x_test.shape
```

```python
x_train.shape
```

# Features Preprocessing

```python
budget_order = ['Low', 'Mid', 'High']
```

## Preprocessor

```python
preprocessor = ColumnTransformer(
    transformers=[
        # Ordinal Encoding for Budget 
        ('budget_ord', OrdinalEncoder(categories=[budget_order]), ['Budget']),
        
        ('text_tfidf', TfidfVectorizer(ngram_range=(1, 3), max_features=2500), 'combined_text')  
    ]
)
```

# Model Creation

```python
model = RandomForestClassifier(
    n_estimators=150,     
    max_depth=12, 
    min_samples_leaf=3,         
    random_state=42       
)
```

## Using Pipeline

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])
```

## Model Training

```python
pipeline.fit(x_train, y_train)
```

# Model Tuning 

## Parameter defining

```python
param_grid = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [10, 15, 20],
    'preprocessor__text_tfidf__max_features': [1500, 2500]
}
```

## Creating GridSearchCV instance

```python
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
```

## Training & searching multiple models

```python
grid_search.fit(x_train, y_train)
```

## Getting best model

```python
best_pipeline = grid_search.best_estimator_
```

## Best score

```python
print(f"Best Accuracy: {grid_search.best_score_}")
```

```python
# Step 1: Input
my_data = pd.DataFrame({
    'Interest': ['IT'],
    'Career Goal': ['Web developer'],
    'Skills': ['coding'],
    'Budget': ['Mid']
})


my_data['combined_text'] = my_data['Interest'] + " " + my_data['Career Goal'] + " " + my_data['Skills']

# Step 2: Predict
prediction = best_pipeline.predict(my_data)

# Step 3: Decode
final_course = le.inverse_transform(prediction)

print(final_course)
```

## Extracting Probabilities 

```python
probs =best_pipeline.predict_proba(my_data)
```

## Decoding Indices into course names

```python
top3_courses = le.inverse_transform(top3_idx)
```

```python
top3_courses
```

## Probabilities of courses

```python
top3_probs
```

## displaying courses with probability
